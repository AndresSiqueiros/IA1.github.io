```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/A1.1 Obesidad.csv')
display(df.head())


```

```python
mean_peso = df['Peso'].mean()
median_peso = df['Peso'].median()
mode_peso = df['Peso'].mode()

print(" Peso")
print("  Media:", mean_peso)
print("  Mediana:", median_peso)
print("  Moda:", mode_peso.to_list())

mean_estatura = df['Estatura'].mean()
median_estatura = df['Estatura'].median()
mode_estatura = df['Estatura'].mode()

print(" Estatura")
print("  Media:", mean_estatura)
print("  Mediana:", median_estatura)
print("  Moda:", mode_estatura.to_list())

mean_agua = df['ConsumoDeAgua'].mean()
median_agua = df['ConsumoDeAgua'].median()
mode_agua = df['ConsumoDeAgua'].mode()

print(" onsumoDeAgua")
print("  Media:", mean_agua)
print("  Mediana:", median_agua)
print("  Moda:", mode_agua)
```

```python
df['IMC'] = df['Peso'] / (df['Estatura']**2)
print("IMC calculado y añadido a la tabla:")
display(df[['Peso', 'Estatura', 'IMC']].head())
```

```python
df_male = df[df['Sexo'] == 'Male']
sns.scatterplot(x='Edad', y='IMC', data=df_male)
plt.title('IMC vs Edad (Hombres)')
plt.xlabel('Edad')
plt.ylabel('IMC')
plt.show()
```

```python
df_male = df[df['Sexo'] == 'Male']
sns.scatterplot(x='Estatura', y='Peso', data=df_male)
plt.title('Estatura vs Peso (Hombres)')
plt.xlabel('Estatura')
plt.ylabel('Peso')
plt.show()
```
